```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
num_pages = 447
all_house_data = []
def chuyen_doi_chuoi_sang_so_nguyen(house_price):
                if 'Triệu' in house_price:
                    return float(house_price.replace(' Triệu', '')) * 0.001
                elif 'Tỷ' in house_price:
                    return float(house_price.replace(' Tỷ', ''))
                else:
                    return 0
txt = ""
for page in range(1, num_pages + 1):
    url = f"https://nhadat24h.net/ban-nha-mat-pho-ha-noi/page{page}"

    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        house_div = soup.find("div", id="ContentPlaceHolder2_KetQuaTimKiem1_Pn1")
        house_container = house_div.find("div", id="divlistitemketquatimkiem_2")
        house_elements = house_container.find_all("div", recursive=False)
        house_data = []

        for house_element in house_elements:
            try:
                 area = house_element.find("label", class_="a-txt-cl2").text.strip()
            except:
                continue
            area = area.replace("m²", "")
            house_price = house_element.find("label", class_="a-txt-cl1").text.replace('m²','').strip()

            price = chuyen_doi_chuoi_sang_so_nguyen(house_price)

            adress_element = house_element.find("i", class_="fas fa-map-marker-alt")
            if adress_element:
                adress_text = adress_element.find_next_sibling(text=True).strip()
            else:
                adress_text = 0


            floor_element = house_element.find("i", class_="fa fa-building")
            if floor_element:
                floor_text = floor_element.find_next_sibling(text=True).strip()
                floor = int(''.join(filter(str.isdigit, floor_text)))
            else:
                floor = 0

            facade_element = house_element.find("i", class_="fa fa-arrows-alt")
            if facade_element:
                facade_text = facade_element.find_next_sibling(text=True).strip()
                facade = int(''.join(filter(str.isdigit, facade_text)))
            else:
                facade = 0

            road_in_element = house_element.find("i", class_="fa fa-road")
            if road_in_element:
                road_in_text = road_in_element.find_next_sibling(text=True).strip()
                road_in = int(''.join(filter(str.isdigit, road_in_text)))
            else:
                road_in = 0


            car_parking_element = house_element.find("i", class_="fa fa-car")
            if car_parking_element:
                car_parking_text = car_parking_element.find_next_sibling(text=True).strip()
                car_parking = int(''.join(filter(str.isdigit, car_parking_text)))
            else:
                car_parking = 0

            bedrooms_element = house_element.find("i", class_="fa fa-bed")
            if bedrooms_element:
                bedrooms_text = bedrooms_element.find_next_sibling(text=True).strip()
                bedrooms = int(''.join(filter(str.isdigit, bedrooms_text)))
            else:
                bedrooms = 0



            wc_element = house_element.find("i", class_="fa fa-bath")
            if wc_element:
                wc_text = wc_element.find_next_sibling(text=True).strip()
                wc = int(''.join(filter(str.isdigit, wc_text)))
            else:
                wc = 0

            house_data.append((area, price, adress_text, floor, facade, road_in, car_parking, bedrooms, wc))
        all_house_data.extend(house_data)
        print(f"Page {page} scraped successfully.")
    else:
        print(f"Failed to scrape page {page}.")
        break
if all_house_data:

    df = pd.DataFrame(all_house_data, columns=['area', 'price', 'Adress', 'floor', 'facade', 'road_in', 'car_parking', 'bedrooms', 'wc'])


    df.to_csv("house_data.csv", index=False, encoding='utf-8-sig')
    print("Dữ liệu đã được lưu vào file house_data_all.xlsx thành công.")
else:
    print("Không tìm thấy thông tin về nhà trên trang web.")
```
